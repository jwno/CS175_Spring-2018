```python
import cv2
import numpy as np
import json
import shutil
import os
from random import randint
```

```python
# replace the path with the path to the annotation.json file
with open('E:\\hand_pose\\Dataset\\annotation.json') as data_file:
    data = json.load(data_file)
```

```python
# Change this path to the location of where the class dataset is
path = 'E:\\hand_pose\\Dataset\\Color\\'

if not os.path.exists("small_dataset"):
    os.makedirs("small_dataset")

for each in data:
    if each[:3] in ['001', '013','025','101','113']:
        #print(each)
        
        min_y = data[each][0][1]
        max_y = data[each][0][1]
        min_x = data[each][0][0]
        max_x = data[each][0][0]

        #print(min_y,min_x)
        for pair in data[each]:
            if pair[1] < min_y:
                min_y = pair[1]
            if pair[1] > max_y:
                max_y = pair[1]
            if pair[0] < min_x:
                min_x = pair[0]
            if pair[0] > max_x:
                max_x = pair[0]

        #print(each)
        file_name = each[:-2]+".jpg"
        full_path = path+file_name
        img = cv2.imread(full_path)

        y_add = 200-(int(max_y)-int(min_y))
        x_add = 200-(int(max_x)-int(min_x))
        
        #y_add = 0
        #x_add = 0

        min_x = int(min_x)
        min_y = int(min_y)
        max_x = int(max_x)
        max_y = int(max_y)

        if y_add % 2 == 0:
            y_cord_min = max(0,min_y - int(y_add/2))
            y_cord_max = min(1080,max_y + int(y_add/2))

        else:
            y_cord_min = max(0,min_y - (1+int(y_add/2)))
            y_cord_max = min(1080,max_y + int(y_add/2))

        if x_add % 2 == 0:
            x_cord_min = max(0,min_x - int(x_add/2))
            x_cord_max = min(1920,max_x + int(x_add/2))

        else:
            x_cord_min = max(0,min_x - (1+int(x_add/2)))
            x_cord_max = min(1920,max_x + int(x_add/2))

        output = "small_dataset\\"+each+".jpg"
        pixels = img[y_cord_min:y_cord_max,x_cord_min:x_cord_max]

        cv2.imwrite(output,pixels)
        cv2.waitKey(0)
        cv2.destroyAllWindows()
```

```python
path = "small_dataset"

if not os.path.exists("right_left_test"):
    os.makedirs("right_left_test")
    os.makedirs("right_left_test\\left")
    os.makedirs("right_left_test\\right")
        
if not os.path.exists("right_left_train"):
    os.makedirs("right_left_train")
    os.makedirs("right_left_train\\left")
    os.makedirs("right_left_train\\right")


for filename in os.listdir(path):
    
    if filename[:3] == "101":
        output =  "right_left_test\\"
    else:
        output =  "right_left_train\\"
    
    if filename[-5] == "L":
        shutil.copy(path+"\\"+filename,output+"left\\"+filename)
    
    elif filename[-5] == "R":
        shutil.copy(path+"\\"+filename,output+"right\\"+filename)
      
    
```

```python
text_file = open("closed_hands.txt","r")

alist = []

for line in text_file:
    alist.append(line.rstrip())

path = "small_dataset"

if not os.path.exists("Opened"):
    os.makedirs("Opened")
        
if not os.path.exists("Closed"):
    os.makedirs("Closed")

for filename in os.listdir(path):
    if filename not in alist:
        shutil.copy(path+"\\"+filename,"Opened\\"+filename)
    else:
        shutil.copy(path+"\\"+filename,"Closed\\"+filename)
    
```

```python

# Directories with Open and Closed File
path_open = r'Opened'
path_close = r'Closed'


# CREATE NEW FOLDER ON PYTHON
# Open Path
path_test_O = r'open_closed_test\open'
path_train_O = r'open_closed_train\open'

# Close Path
path_test_C = r'open_closed_test\closed'
path_train_C = r'open_closed_train\closed'

# Percentage(%)
test = 20


if not os.path.exists(path_test_O):
    os.makedirs(path_test_O)
        
if not os.path.exists(path_train_O):
    os.makedirs(path_train_O)

if not os.path.exists(path_test_C):
    os.makedirs(path_test_C)
        
if not os.path.exists(path_train_C):
    os.makedirs(path_train_C)
    

paths = [path_open, path_close]
path_list = [[path_test_O, path_train_O,], [path_test_C, path_train_C,]]
for i, path in enumerate(paths):
    list_path = []
    for filename in os.listdir(path):
        list_path.append(filename)

    takeout = int(len(list_path)*(test*0.01))
    for j in range(takeout):
        random_number = randint(0, len(list_path)-1)
        newpath = os.path.join(path, list_path[random_number])
        shutil.copy(newpath, path_list[i][0])
        del list_path[random_number]

    for j in list_path:
        newpath = os.path.join(path, j)
        shutil.copy(newpath, path_list[i][1])
```
